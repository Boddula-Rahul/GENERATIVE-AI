```python
from langchain.prompts.chat import ChatPromptTemplate
```

```python
from langchain.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate
```

```python
system_prompt = SystemMessagePromptTemplate.from_template("You are a helpful assitant who answers questions clearly along with simple explanation")
```

```python
system_prompt
```

```python
human_prompt = HumanMessagePromptTemplate.from_template("Question : {user_question}")
```

```python
human_prompt
```

```python
chat_prompt = ChatPromptTemplate([system_prompt,human_prompt])
```

```python
chat_prompt
```

```python
format_msg= chat_prompt.format_messages(user_question = "What is the Capital of INDIA")
```
