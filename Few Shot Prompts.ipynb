```python
from langchain.prompts import FewShotPromptTemplate, PromptTemplate
```

**Examples*

```python
examples = [
             {"input" : "2 + 2",
              "output" : "4"
              },
             {"input " : " 3 * 3",
              "output" : "9"
              }
]
```

```python
examples
```

```python
eg_prompt = PromptTemplate(

               input_variables = ["input", "output"] ,
               template = "Input : {input} Output : {output}"
               )
```

```python
FS_prompt = FewShotPromptTemplate(

                      examples = examples,
                      example_prompt = eg_prompt,
                      prefix = "Solve following math problem",
                      suffix =  "Input : {input} Output : ",
                      input_variables = ["input"]

                      )
```

```python
format_pt = FS_prompt.format(input = "5 -2")
```
