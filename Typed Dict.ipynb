```python
from typing import TypedDict
```

```python
class book(TypedDict) :
    title :str
    author : str
    pages : int
    price : float



new_book : book = {"title" : "Night in Woods",
        "author" : "Rahul",
        "pages" : 534,
        "price" : 352.75
        }
```

```python
print(new_book)
```

```python
new_book1 : book = {"title" : "Night in Woods",
        "author" : "Rahul",
        "pages" : "small",
        "price" : "In Rupees"
        }
```

```python
print(new_book1)
```
