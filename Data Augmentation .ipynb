```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
imgen = ImageDataGenerator(

                rotation_range=45,
                width_shift_range=0.20,
                height_shift_range=0.20,
                shear_range=0.20,
                zoom_range=0.20,
                horizontal_flip=True,
                vertical_flip=True,
                fill_mode='nearest')
```
