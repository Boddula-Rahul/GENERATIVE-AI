## **Assigning Default Value **

```python
from pydantic import BaseModel
```

```python
class student(BaseModel):
  name : str = "Sai" #Default value set up

new_std1 = {}
std = student(**new_std1)
print(std)
```

## Optional

```python
from pydantic import BaseModel
from typing import Optional
```

```python
class product(BaseModel) :
  prod : Optional[str] = "laptop"
  quantity : int = 10
  price : Optional[int] = None


prod1 = {}
laptop =  product(**prod1)
print(laptop)

```

```python
prod2 = {"price" : 50000}

laptop2 = product(**prod2)
print(laptop2)
```

```python
prod2 = {"prod" : "apple", "price" : 50000}

laptop2 = product(**prod2)
print(laptop2)
```
